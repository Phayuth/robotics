# Single

```python
path = "/home/yuth/ws_yuthdev/robotics_manipulator/datasave/new_data/"
# 0. RRTBase
# 1. RRTStar
# 2. RRTInformed
# 3. RRTStarQuick

# 4. RRTConnect
# 5. RRTStarConnect
# 6. RRTInformedConnect
# 7. RRTStarConnectQuick

# 8. RRTBaseMulti
# 9. RRTStarMulti
# 10. RRTInformedMulti
# 11. RRTStarQuickMulti

# 12. RRTConnectMulti
# 13. RRTStarConnectMulti
# 14. RRTStarConnectQuickMulti

# 15. RRTConnectNoneAgressive - baseline
# 16. RRTConnectNoneAgressiveMulti - baseline

a = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_2_withlocgap_False_performance.csv").drop('Unnamed: 0', axis=1)
b = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_4_withlocgap_False_performance.csv").drop('Unnamed: 0', axis=1)
c = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_15_withlocgap_False_performance.csv").drop('Unnamed: 0', axis=1)
d = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_5_withlocgap_False_performance.csv").drop('Unnamed: 0', axis=1)
e = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_6_withlocgap_False_performance.csv").drop('Unnamed: 0', axis=1)

f = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_4_withlocgap_True_performance.csv").drop('Unnamed: 0', axis=1)
g = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_15_withlocgap_True_performance.csv").drop('Unnamed: 0', axis=1)
h = pd.read_csv(path + "env_RobotArm2DSimulator_types_Single_planner_5_withlocgap_True_performance.csv").drop('Unnamed: 0', axis=1)
```

# Mean

```python
am = a.mean(axis=0)
bm = b.mean(axis=0)
cm = c.mean(axis=0)
dm = d.mean(axis=0)
em = e.mean(axis=0)
fm = f.mean(axis=0)
gm = g.mean(axis=0)
hm = h.mean(axis=0)

df = pd.concat([am, bm, cm, dm, em, fm, gm, hm], axis=1)
df.rename(
    columns={
        0: "inform",
        1: "connect",
        2: "connect na",
        3: "star connect",
        4: "infrom star connect",
        5: "connect locgap",
        6: "connect na locgap",
        7: "star connect locgap",
    },
    inplace=True,
)
df.T
```

# STD

```python
astd = a.std(axis=0)
bstd = b.std(axis=0)
cstd = c.std(axis=0)
dstd = d.std(axis=0)
estd = e.std(axis=0)
fstd = f.std(axis=0)
gstd = g.std(axis=0)
hstd = h.std(axis=0)

dfs = pd.concat([astd, bstd, cstd, dstd, estd, fstd, gstd, hstd], axis=1)
dfs.rename(
    columns={
        0: "inform",
        1: "connect",
        2: "connect na",
        3: "star connect",
        4: "infrom star connect",
        5: "connect locgap",
        6: "connect na locgap",
        7: "star connect locgap",
    },
    inplace=True,
)
dfs.T
```

# To thesis

```python
df = df.T
dfs = dfs.T
df = df.round(2)
dfs = dfs.round(2)
```

```python
dfnew = df.copy()
dfnew['First Cost'] = '&$' + df['First Cost'].map(str) + ' \pm ' + dfs['First Cost'].map(str) + '$'
dfnew['First Iteration'] = '&$' + df['First Iteration'].map(str) + ' \pm ' + dfs['First Iteration'].map(str) + '$'
dfnew['Latest Cost'] = '&$' + df['Latest Cost'].map(str) + ' \pm ' + dfs['Latest Cost'].map(str)+ '$'
dfnew['Latest Iteration'] = '&$' + df['Latest Iteration'].map(str) + ' \pm ' + dfs['Latest Iteration'].map(str)+ '$'
dfnew['Run Time'] = '&$' + df['Run Time'].map(str) + ' \pm ' + dfs['Run Time'].map(str)+ '$'
dfnew['Plan Time'] = '&$' + df['Plan Time'].map(str) + ' \pm ' + dfs['Plan Time'].map(str)+ '$'
dfnew['Col.C Time'] = '&$' + df['Col.C Time'].map(str) + ' \pm ' + dfs['Col.C Time'].map(str)+ '$'
dfnew['Node Number'] = '&$' + df['Node Number'].map(str) + ' \pm ' + dfs['Node Number'].map(str)+ '$'
dfnew['Col.C Number'] = '&$' + df['Col.C Number'].map(str) + ' \pm ' + dfs['Col.C Number'].map(str)+ '$'
dfnew
```
